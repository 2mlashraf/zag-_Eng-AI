### **2.Find the 5 states with the lowest and highest `crime rates`**

```python
# find the 5 states with the lowest crime rates
print(df.columns)
lowest_crime_states = df.nsmallest(5, 'Crime Rate')
print(lowest_crime_states)
```

```python
import matplotlib.pyplot as plt

# Create a bar plot to visualize the 5 states with the lowest crime rate
lowest_crime_states = df.nsmallest(5, 'Crime Rate')


# Set title and labels for the plot

plt.figure(figsize=(10, 5))
plt.bar(lowest_crime_states['State'], lowest_crime_states['Crime Rate'], color='green')
plt.xlabel('State')
plt.ylabel('Crime Rate')
plt.title('Top 5 States with Lowest Crime Rate')

plt.xticks(rotation=45)


# Display the plot
plt.show()

# Print your conclusions
print("Conclusion: ")
```

```python
# get the 5 states with the highest crime rates.
top_5_states = df.sort_values(by='Crime Rate', ascending=False).head(5)
```

```python
# Create a bar plot to visualize the 5 states with the Highest crime rate
plt.figure(figsize=(10, 6))
plt.bar(top_5_states['State'], top_5_states['Crime Rate'], color='red')

# Set title and labels for the plot

plt.title('Top 5 States with Highest Crime Rate', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Crime Rate', fontsize=12)
# Display the plot
plt.show()

# Print your conclusions
print("Conclusion: ")
```

### **3.Find the sum of `Rape Cases` for each year**

```python
# Group the dataset by 'Year' and calculate the sum of 'Rape Cases' for each year
yearly_cases = df.groupby('Year')['Rape Cases'].sum()

# Plot the line chart showing the trend of total rape cases per year
plt.figure(figsize=(10, 6))
plt.plot(yearly_cases.index, yearly_cases.values, marker='o', linestyle='-', color='red')


# Title and labels for the plot
plt.title('Trend of Rape Cases per Year', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Rape Cases', fontsize=12)
# Show the plot
plt.show()

# Print your conclusions from this plot
print("conclusions: ")
```

### **4.Find the correlation between `total crimes` and `population`**

```python
# Create a scatter plot with Population on the x-axis and Total Crimes on the y-axis
plt.figure(figsize=(10, 6))
plt.scatter(df['Population'], df['Total Crimes'], color='blue', alpha=0.6)
plt.grid(True)

# Set title and labels for the plot
plt.title('Correlation between Population and Total Crimes', fontsize=14)
plt.xlabel('Population', fontsize=12)
plt.ylabel('Total Crimes', fontsize=12)


# Show the plot
plt.show()

# Print your conclusions from this plot
print("conclusions: ")
```

### **5.Find out how the average of `crime clearance` rate vary across different `states`**

```python
# Group the data by 'State' and calculate the mean 'Crime Clearance Rate' for each state, then sort them in descending order
state_clearance_rates = df.groupby('State')['Crime Clearance Rate'].mean().sort_values(ascending=False)

# Create a bar plot using the calculated average crime clearance rates
plt.figure(figsize=(12, 6))
plt.bar(state_clearance_rates.index, state_clearance_rates.values, color='orange')


# Set the title and labels for the plot
plt.title('Average Crime Clearance Rate per State', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Average Crime Clearance Rate', fontsize=12)


# Rotate x-axis labels (state names) to avoid overlap and ensure readability
plt.xticks(rotation=45, ha="right")


# Show the plot
plt.show()
# Print your conclusions from this plot
print("conclusions: ")

```

### **7.Find the trend in the number of `murder cases` over the `years`**

```python
# get the sum of the murder cases over the years
murder_trend = df.groupby('Year')['Murder Cases'].sum()


# Create a line plot with murder cases trend
plt.figure(figsize=(10, 5))
plt.plot(murder_trend.index, murder_trend.values, marker='o', linestyle='-', color='red')


# Set the title and labels for the plot
plt.title('Trend of Murder Cases Over the Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Murder Cases', fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
# Print your conclusions from this plot
print("conclusions: ")
```

### **8.Find out which `state` has highest number of `arrest`**

```python
#pick the top 10 states
arrest_by_state = df.groupby('State')['Arrests'].sum()

#barplot for the top 10 states
top_10_states = arrest_by_state.sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_10_states.index, top_10_states.values, color='skyblue')


# Set the title and labels
plt.title('Top 10 States with Highest Number of Arrests', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Total Arrests', fontsize=12)
plt.xticks(rotation=45)

# Show the plot
plt.show()

# Print your conclusions from this plot
print("conclusions: ")

```

### **9.Find the average `crime rate` across southern states over the `years`**

Note: southern states are 'Andhra Pradesh', 'Telangana', 'Tamil Nadu', 'Karnataka', 'Kerala', 'Puducherry'

```python
#Southern States

southern_states = ['Andhra Pradesh', 'Telangana', 'Tamil Nadu', 'Karnataka', 'Kerala', 'Puducherry']
# Filter the dataset for Southern States
southern_data = df[df['State'].isin(southern_states)]


# Calculate the average crime rate by state and year
avg_crime_rate = southern_data.groupby('Year')['Crime Rate'].mean()

# Plot the line plot for Southern States

plt.figure(figsize=(10, 5))
plt.plot(avg_crime_rate.index, avg_crime_rate.values, marker='o', linestyle='-', color='b')

# Set title and labels
plt.title('Average Crime Rate in Southern States Over the Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Crime Rate', fontsize=12)
plt.grid(True)

# Show the plot
plt.show()

# Print your conclusions from this plot
print("conclusions: ")

```

### **10.Find the Total number of `cases` for each crime type**

Note: cases are Rape cases, Abduction cases, Murder cases

```python
# Sum the cases for each crime type
crime_type_counts = df[['Rape Cases', 'Abduction Cases', 'Murder Cases']].sum()

# Plot the donut chart
plt.figure(figsize=(7, 7))
plt.pie(crime_type_counts, labels=crime_type_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'white', 'linewidth': 2})
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)

# Set the title
plt.title('Total Number of Cases for Each Crime Type')


# Show the plot
plt.show()

# Print your conclusions from this plot
print("conclusions: ")
```

### **12.Find the Correlation Matrix of Crime Data**

```python
# Filter data to ensure only numeric data is left

numeric_df = df.select_dtypes(include=['number'])
# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Set up the visual style of the plot
plt.figure(figsize=(10, 8))
sns.set(style="whitegrid")

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)


# Set the title
plt.title("Correlation Matrix of Crime Data")


# Display the plot
plt.show()
# Print your conclusions from this plot
print("conclusions: ")
```
