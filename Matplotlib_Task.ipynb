# Matplotlib Practice

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import the 'data/car-sales.csv' into a DataFame called car_sales and view
car_sales = pd.read_csv("car-sales.csv")
car_sales.head()
```

```python
# Try to plot the 'Price' column using the plot() function
car_sales["Price"].plot(
    kind='line',
    figsize=(10, 5),
    title='Car Prices',
    ylabel='Price ($)',
    xlabel='Car Index',
    color='green',
    marker='o'
    )
plt.grid(True)
plt.show()
```

Why doesn't it work?

Hint: It's not numeric data.

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
car_sales['Price'] = car_sales['Price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
car_sales['Total Sales'] = car_sales['Price'].cumsum()

# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
car_sales['Sale Date'] = pd.date_range(start='2025-05-12', periods=len(car_sales), freq='D')


# View the car_sales DataFrame
print(car_sales)
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column

plt.figure(figsize=(10, 6))

car_sales.plot(
    x='Sale Date',
    y='Total Sales',
    kind='line',
    marker='o',
    linestyle='-',
    color='royalblue',
    linewidth=2,
    markersize=8,
    ax=plt.gca()
)
plt.title('Cumulative Car Sales Revenue', fontsize=14, pad=20)
plt.xlabel('Sale Date', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Rotate x-axis labels
plt.xticks(rotation=45)

plt.show()
```

```python
# Convert the 'Price' column to the integers
car_sales['Price'] = car_sales['Price'].astype(int)
print(car_sales[['Price']].head())

# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function
car_sales.plot(x='Odometer (KM)', y='Price', kind='scatter', figsize=(10,6))

plt.title('Odometer vs Price')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
# Create a NumPy array of random numbers of size (10, 4) and save it to X
X = np.random.rand(10, 4)
print("Random array X (10 rows, 4 columns):")
print(X)

# Turn the NumPy array X into a DataFrame with columns called ['a', 'b', 'c', 'd']
X_df = pd.DataFrame(X, columns=['a', 'b', 'c', 'd'])
print(X_df.head())

# Create a bar graph of the DataFrame
X_df.plot(kind='bar')

plt.title('Bar Chart of Each Row in DataFrame')
plt.xlabel('Row Index')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
# Create a bar graph of the 'Make' and 'Odometer (KM)' columns in the car_sales DataFrame

plt.figure(figsize=(10, 6))

car_sales.plot(
    x='Make',
    y='Odometer (KM)',
    kind='bar',
    color='lightgreen',
    edgecolor='black',
    width=0.7
)

plt.title('Car Mileage by Brand', fontsize=14, pad=20)
plt.xlabel('Car Make', fontsize=12)
plt.ylabel('Odometer (KM)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of each bar
for index, value in enumerate(car_sales['Odometer (KM)']):
    plt.text(
        index,
        value + 1000,
        f'{value:,} KM',  # Format with thousands separator
        ha='center',
        va='bottom',
        fontsize=7
    )

plt.tight_layout()
plt.show()

```

```python
# Create a histogram of the 'Odometer (KM)' column

plt.hist(car_sales['Odometer (KM)'],
         bins=5,
         color='skyblue',
         edgecolor='black',
         alpha=0.7)


# Add mean line with legend
mean_value = car_sales['Odometer (KM)'].mean()
plt.axvline(mean_value,
            color='red',
            linestyle='--',
            label=f'Mean: {mean_value:.0f} KM')

plt.legend()
plt.title('Odometer (KM) Distribution')
plt.xlabel('Odometer (KM)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a histogram of the 'Price' column with 20 bins

if 'Price' not in car_sales.columns:
    raise ValueError("DataFrame doesn't contain a 'Price' column")

print("Original Price column sample:")
print(car_sales['Price'].head())

# Clean the Price data
try:
    # Convert to string
    car_sales['Price'] = car_sales['Price'].astype(str)

    # Remove $ and commas, convert to float
    car_sales['Price_clean'] = (
        car_sales['Price']
        .str.replace('[\$,]', '', regex=True)
        .astype(float)
    )

    print("\nCleaned Price values sample:")
    print(car_sales['Price_clean'].head())

except Exception as e:
    print(f"Error cleaning Price data: {e}")
    raise

if 'Price_clean' in car_sales.columns:
    plt.figure(figsize=(12, 6))
    plt.hist(car_sales['Price_clean'], bins=20, color='skyblue', edgecolor='black')

    plt.title('Car Price Distribution (20 Bins)')
    plt.xlabel('Price ($)')
    plt.ylabel('Number of Cars')
    plt.grid(axis='y', alpha=0.3)

    # mean and median lines
    mean_price = car_sales['Price_clean'].mean()
    median_price = car_sales['Price_clean'].median()

    plt.axvline(mean_price, color='red', linestyle='--', label=f'Mean: ${mean_price:,.2f}')
    plt.axvline(median_price, color='green', linestyle=':', label=f'Median: ${median_price:,.2f}')
    plt.legend()

    plt.show()
else:
    print("Could not create histogram - price cleaning failed")
```

```python
# View the first 10 rows of the heart_disease DataFrame

print(heart_disease.head(10))
```

```python
# Create a histogram of the "age" column with 50 bins


plt.figure(figsize=(14, 7))

plt.hist(heart_disease['age'],
         bins=50,
         color='skyblue',
         edgecolor='blue',
         alpha=0.85)

plt.title('Age Distribution of Heart Disease Patients\n(50 Bins)',
          fontsize=16, pad=20)
plt.xlabel('Age (Years)', fontsize=14)
plt.ylabel('Number of Patients', fontsize=14)

mean_age = heart_disease['age'].mean()
median_age = heart_disease['age'].median()
plt.axvline(mean_age, color='red', linestyle='--',
            linewidth=1.5, label=f'Mean: {mean_age:.1f} yrs')
plt.axvline(median_age, color='green', linestyle='--',
            linewidth=1.5, label=f'Median: {median_age:.1f} yrs')

plt.grid(axis='y', linestyle=':', alpha=0.4)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Call plot.hist() on the heart_disease DataFrame and toggle the
# "subplots" parameter to True

heart_disease.plot.hist(
    subplots=True,
    figsize=(10, 15),
    bins=20,
    edgecolor='white',
    alpha=0.7,
    layout=(-1, 2),
    legend=False
)

plt.tight_layout()
plt.show()
```

That plot looks pretty squished. Let's change the figsize.

```python
# Call the same line of code from above except change the "figsize" parameter
# to be (10, 30)

numeric_cols = heart_disease.select_dtypes(include=['number']).columns

# Create figure and axes for full control
fig, axes = plt.subplots(
    len(numeric_cols), 1,          # One subplot per numeric column
    figsize=(12, 6*len(numeric_cols))  # Dynamic height based on column count
)


for ax, col in zip(axes, numeric_cols):
    heart_disease[col].plot.hist(
        ax=ax,
        bins=20,
        color='steelblue',
        edgecolor='white',
        title=f'Distribution of {col}'
    )
    ax.grid(alpha=0.3)
    ax.set_xlabel('')  # Cleaner look without repeated x-labels

plt.tight_layout(h_pad=4.0)
plt.show()
```
