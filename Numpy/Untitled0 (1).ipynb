```python
import numpy as np

#create a ndarray [10, 10, 3] with all elements set to one.
x = np.ones((10, 10, 3))

#  Reshape x to have the second dimension equal to 150
# 10 * 10 * 3 = 300, so the new shape must be 300
new_shape = (2, 150, 1)  # The product is 2 * 150 * 1 = 300
reshaped_x = x.reshape(new_shape)

# Display the reshaped arr
print(reshaped_x)
print("New shape:", reshaped_x.shape)
```
